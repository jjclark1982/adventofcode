# [Day 1: Historian Hysteria](https://adventofcode.com/2024/day/1)

```python
import aocd
input_data = aocd.get_data(year=2024, day=1).split('\n')
input_data[:5]
```

---

### Part 1

```python
import pandas as pd
```

```python
df = pd.DataFrame(((int(s) for s in d.split()) for d in input_data), columns=['left', 'right'])
df
```

```python
df['left'][:] = df['left'].sort_values()
```

```python
df['right'][:] = df['right'].sort_values()
```

```python
df['distance'] = (df['left'] - df['right']).abs()
```

```python
df
```

#### Part 1 Answer

**What is the total distance between your lists?**

```python
df['distance'].sum()
```

---
### Part 2

Calculate a total similarity score by adding up each number in the left list after multiplying it by the number of times that number appears in the right list.

```python
l = df['left'].to_numpy() # axis 0
r = df['right'].to_numpy().reshape(-1, 1) # axis 1
num_similar = (l == r).sum(axis=0)
```

```python
df['num_similar'] = num_similar
df[df['num_similar'] != 0]
```

#### Part 2 Answer

Once again consider your left and right lists. **What is their similarity score?**

```python
(df['num_similar'] * l).sum()
```
